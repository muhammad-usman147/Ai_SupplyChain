### Creating DataSet For Supply Chain Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
apr = pd.read_csv("PK COVID-19-30apr.csv")
apr
```

```python
may = pd.read_csv("PK COVID-19-10may.csv")
may
```

```python
june = pd.read_csv("PK COVID-19-3jun.csv")
june
```

```python
pd.concat([apr,may])
```

```python
df = pd.concat([pd.concat([apr,may]),june])
df
```

```python
y = list(df['Cases'])
```

```python
x = list(df['Date'])
```

```python
figure = plt.figure(figsize=(20,10))
plt.xticks(rotation = 90)
sns.lineplot(x = x, y = y)
```

```python
figure = plt.figure(figsize=(20,10))
plt.xticks(rotation = 90)
sns.barplot(x = x, y = y)
```

```python
np.random.seed(8)
masks_sold = [np.random.randint(500,1000) for i in range(len(x))]
```

```python
figure = plt.figure(figsize=(20,10))
plt.xticks(rotation = 90)
plt.title("Masks sold during Covid-19")
sns.lineplot(x = x, y = masks_sold)
```

```python
np.random.seed(9)
sanitizers = [np.random.randint(500,5000) for i in range(len(x))]
```

```python
figure = plt.figure(figsize=(20,10))
plt.xticks(rotation = 90)
plt.title("Sanitizers sold during Covid-19")
sns.lineplot(x = x, y = sanitizers)
```
